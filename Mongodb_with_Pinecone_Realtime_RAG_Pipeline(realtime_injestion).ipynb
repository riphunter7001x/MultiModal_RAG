```python
from pinecone import Pinecone

pc = Pinecone(api_key="f26cb429-8611-49f7-9df1-ca73f1b3a098")
index = pc.Index("moviesdb")
```

```python

def get_result(query,similar_result=3):
  embedding=embedding_model.encode(query)
  embedding=embedding.tolist()

  result=index.query(
    vector=embedding,
    top_k=similar_result,
  )
  return result
```

```python

query="what is the best comedy movie to watch and why?"
```

```python

result=get_result(query)
```

```python
result
```

```python

from bson.objectid import ObjectId
```

```python

mylist=[]
for i in  range(len(result["matches"])):
  value=result["matches"][i]['id']
  mylist.append(collection.find_one({"_id": ObjectId(value)}))
```

```python
mylist
```

```python
combined_information = ""
for i in range(len(mylist)):
  fullplot=mylist[i]["fullplot"]
  title=mylist[i]["title"]
  imdb = mylist[i]["imdb"]["rating"]
  combined_information += f"Title:{title}, imdb_rating {imdb}, fullplot: {fullplot}\n"
```

```python

print(combined_information)
```
